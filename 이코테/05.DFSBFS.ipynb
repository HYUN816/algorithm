```python
# 음료수 얼려먹기 

# connected component 찾기 

# DFS

# 알고리즘(강의)
# 1. 특정한 지점의 주변 상,하,좌,우를 살펴본 뒤에 주변 지점 중에서 값이 '0'이면서 아직 방문하지 않은 지점이 있다면 해당 지점을 방문
# 2. 방문한 지점에서 다시 상,하,좌,우를 살펴보면서 방문을 진행하는 과정 반복 -> 연결된 모든 지점 방문 가능
# 3. 모든 노드에 대하여 1~2번 과정을 반복, 방문하지 않은 지점의 수를 카운트

def dfs(x,y) :
    print('dfs함수',x,y)
    # 주어진 범위를 벗어나는 경우에는 즉시 종료 
    if x<=-1 or x>= n or y<=-1 or y>=m :
        print('out of range')
        return False
    
    if graph[x][y]==0 :
        print('방문처리',x,y)
        #해당노드 방문 처리 
        graph[x][y]=1
        
        dfs(x-1,y)
        dfs(x,y-1)
        dfs(x+1,y)
        dfs(x,y+1)
        
        return True
    return False
```

```python
# 4 5
# 00110
# 00011
# 11111
# 00000



n,m = map(int,input().split())

graph=[]

for i in range(n):
    graph.append(list(map(int,input())))
    
result = 0 

for i in range(n):
    for j in range(m):
        if dfs(i,j)==True:
            print(i,j)
            result+=1
print('result : ',result)
```

```python
# 미로탈출 

# BFS 이용
# 시작지점에서 가까운 노드부터 차례대로 그래프의 모든 노드를 탐색 

# 알고리즘(강의)
# 1. (1,1)위치에서 시작
# 2. (1,1)좌표에서 상하좌우로 탐색을 진행하면 바로 옆 노드인(1,2) 위치의 노드를 방문하게되고 새롭게 방문하는 (1,2)노드의 값을 2로 바꾸게 된다
# 3. 마찬가지로 BFS를 계속 수행하면 결과적으로 최단경로의 값들이 1씩 증가하는 형태로 변경된다. 

def bfs(x,y) :
    queue = deque()
    queue.append((x,y))

    # 큐가 빌 때까지 반복하기 
    while queue : 
        print('queue',queue)
        x, y = queue.popleft()
        print('x',x, 'y',y)
        #현재 위치에서 4가지 방향으로의 위치 확인
        for i in range(4):
            nx = x +dx[i]
            ny = y +dy[i]
            print('nx',nx,'ny',ny)
            
            #미로찾기 공간을 벗어난 경우 무시
            if nx < 0 or nx >= n or ny < 0 or ny >= m :
                print('out of range')
                continue
                
            #벽인경우 무시
            if graph[nx][ny] ==0 :
                print('wall')
                continue
                
            # 해당 노드를 처음 방문하는 경우에만 최단 거리 기록
            if graph[nx][ny] == 1 : 
                graph[nx][ny] = graph[x][y] +1
                queue.append((nx,ny))
                print('visited', nx,ny,graph[nx][ny])
                
    #가장 오른쪽 아래까지의 최단 거리 반환
    return graph[n-1][m-1]
```

```python
# 5 6
# 101010
# 111111
# 000001
# 111111
# 111111

from collections import deque

n,m = map(int, input().split())
graph=[]

for i in range(n):
    graph.append(list(map(int,input())))
    
# 이동할 네가지 방향 정의 (상,하,좌,우)
dx=[-1,1,0,0]
dy=[0,0,-1,1]

print(bfs(0,0))
```
